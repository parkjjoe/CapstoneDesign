이미지가 복사되어 train과 valid로 나뉘어짐.

```python
import os
import random
import shutil

# 데이터셋 경로 설정
data_path = "/content/drive/MyDrive/data/"

# 학습 데이터셋과 검증 데이터셋을 저장할 경로 설정
train_path = "/content/drive/MyDrive/data/train/"
valid_path = "/content/drive/MyDrive/data/val/"

# 학습 비율 설정
train_ratio = 0.8

# 데이터셋 폴더 안의 이미지 파일 목록 가져오기
file_list = os.listdir(data_path)
file_list = [file for file in file_list if file.endswith('.jpg')]

# 이미지 파일 목록을 무작위로 섞음
random.shuffle(file_list)

# 학습 데이터셋과 검증 데이터셋으로 나누기
train_size = int(len(file_list) * train_ratio)
train_files = file_list[:train_size]
valid_files = file_list[train_size:]

# 학습 데이터셋 폴더 만들기
os.makedirs(train_path, exist_ok=True)

# 검증 데이터셋 폴더 만들기
os.makedirs(valid_path, exist_ok=True)

# 학습 데이터셋으로 이미지 파일 이동
for file in train_files:
    src_path = os.path.join(data_path, file)
    dst_path = os.path.join(train_path, file)
    shutil.copy(src_path, dst_path)

# 검증 데이터셋으로 이미지 파일 이동
for file in valid_files:
    src_path = os.path.join(data_path, file)
    dst_path = os.path.join(valid_path, file)
    shutil.copy(src_path, dst_path)
```

train:valid:test = 7:2:1의 비율로 이미지를 이동시킴.

```python
import os
import random

# 데이터셋 경로 설정
data_path = "/content/drive/MyDrive/data_G/right_chopping/"

# 학습 데이터셋과 검증 데이터셋, 테스트 데이터셋을 저장할 경로 설정
train_path = "/content/drive/MyDrive/data_G/train/right_chopping/"
valid_path = "/content/drive/MyDrive/data_G/val/right_chopping/"
test_path = "/content/drive/MyDrive/data_G/test/right_chopping/"

# 데이터셋을 나눌 비율 설정
train_ratio = 0.7
valid_ratio = 0.2
test_ratio = 0.1

# 데이터셋 폴더 안의 이미지 파일 목록 가져오기
file_list = os.listdir(data_path)
file_list = [file for file in file_list if file.endswith('.jpg')]

# 이미지 파일 목록을 무작위로 섞음
random.shuffle(file_list)

# 학습 데이터셋과 검증 데이터셋, 테스트 데이터셋으로 나누기
train_size = int(len(file_list) * train_ratio)
valid_size = int(len(file_list) * valid_ratio)
test_size = int(len(file_list) * test_ratio)

train_files = file_list[:train_size]
valid_files = file_list[train_size:train_size+valid_size]
test_files = file_list[train_size+valid_size:]

# 데이터셋 폴더 만들기
os.makedirs(train_path, exist_ok=True)
os.makedirs(valid_path, exist_ok=True)
os.makedirs(test_path, exist_ok=True)

# 학습 데이터셋으로 이미지 파일 이동
for file in train_files:
    src_path = os.path.join(data_path, file)
    dst_path = os.path.join(train_path, file)
    os.rename(src_path, dst_path)

# 검증 데이터셋으로 이미지 파일 이동
for file in valid_files:
    src_path = os.path.join(data_path, file)
    dst_path = os.path.join(valid_path, file)
    os.rename(src_path, dst_path)

# 테스트 데이터셋으로 이미지 파일 이동
for file in test_files:
    src_path = os.path.join(data_path, file)
    dst_path = os.path.join(test_path, file)
    os.rename(src_path, dst_path)
```
