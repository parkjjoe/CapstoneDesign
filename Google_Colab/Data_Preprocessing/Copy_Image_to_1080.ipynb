```python
import os
from PIL import Image

# 원본 데이터 경로와 저장할 새 데이터 경로
original_dataset_dir = '/content/drive/MyDrive/data_FG_10Class'
new_dataset_dir = '/content/drive/MyDrive/data_FG_10Class_1080_2'

# test, train, val 폴더에 대해서 반복
for set_name in ['test', 'train', 'val']:
    original_set_dir = os.path.join(original_dataset_dir, set_name)
    new_set_dir = os.path.join(new_dataset_dir, set_name)

    # 각 클래스에 대해서 반복
    for class_name in os.listdir(original_set_dir):
        original_class_dir = os.path.join(original_set_dir, class_name)
        new_class_dir = os.path.join(new_set_dir, class_name)
        os.makedirs(new_class_dir, exist_ok=True)

        # 각 이미지에 대해서 반복
        for img_name in os.listdir(original_class_dir):
            original_img_path = os.path.join(original_class_dir, img_name)
            new_img_path = os.path.join(new_class_dir, img_name)

            # 이미지를 로드하고 크롭
            img = Image.open(original_img_path)
            width, height = img.size

            new_img = img.crop((0, 0, 1080, 1080))  # 이미지의 상단을 기준으로 1080x1080으로 크롭

            # 이미지를 새 위치에 저장
            new_img.save(new_img_path)
```
