```python
import cv2 #opencv 불러오기

video_cap=cv2.VideoCapture('/content/drive/MyDrive/video_G/yang_video_G/back_right_hurray.mp4') #영상파일 불러와서 캡쳐

count = 0 #이미지 이름정하기용 변수

while(video_cap.isOpened()): #정상적으로 영상이 열렸다면
    ret, image = video_cap.read() #캡쳐해서 변수에 저장 ret은 성공/실패 여부 , image에 캡쳐한것 프레임단위로 입력
    if image is None: #만약 캡쳐가 안됐다면, 즉 추출이 끝났다면
      print("추출 끝")
      break
    else:
      img = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE) # 이미지 90도 회전

    if(int(video_cap.get(cv2.CAP_PROP_POS_FRAMES)) % 1 == 0): #get으로 영상의 현재 프레임을 받아 1프레임 단위로 추출
        print('추출중인 프레임 :'+ str(int(video_cap.get(cv2.CAP_PROP_POS_FRAMES)))) #추출 프레임 텍스트로 표현
        cv2.imwrite("/content/drive/MyDrive/data_G/back_right_hurray/back_right_hurray_Gy%d.jpg"%count,img) #이미지파일을 저장
        count+=1 #넘버링+1

print("영상의 초당 frame은 " + str(video_cap.get(cv2.CAP_PROP_FPS)) + " frame")  # 초당 몇프레임의 영상인지 확인

video_cap.release() #열린 영상 닫기.
```

```python
import os

def count_images(folder_path, extensions=['.jpg']):
    image_count = 0
    for file in os.listdir(folder_path):
        if file.lower().endswith(tuple(extensions)):
            image_count += 1
    return image_count

folder_path = '/content/drive/MyDrive/data_FG_10Class_1080_2/val/front_left_raised'
image_count = count_images(folder_path)
print(f"폴더 내 이미지 수: {image_count}")
```

```python
import os

dir_path = '/content/drive/MyDrive/image/front'
file_names = os.listdir(dir_path)

for i, file_name in enumerate(file_names):
    ext = os.path.splitext(file_name)[-1] # 파일 이름에서 확장자 추출
    new_file_name = f'front{i}{ext}' # 새 파일 이름 생성
    os.rename(os.path.join(dir_path, file_name), os.path.join(dir_path, new_file_name)) # 파일 이름 변경
```
