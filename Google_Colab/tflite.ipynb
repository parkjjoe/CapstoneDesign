```python
import tensorflow as tf

model = tf.keras.models.load_model('/content/drive/MyDrive/ResNet50_DEFG_ImageDataGenerator.h5')

converter = tf.lite.TFLiteConverter.from_keras_model(model)

tflite_model = converter.convert()

with open('ResNet50_DEFG_ImageDataGenerator.tflite', 'wb') as f:
    f.write(tflite_model)
```
